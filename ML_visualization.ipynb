```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_predictions(csv_file, output_dir='ML_plots'):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Extract the real and predicted values
    real_values = df['Real']
    predicted_values = df['Predicted']
    
    # Create a plot
    plt.figure(figsize=(10, 6))
    plt.scatter(real_values, predicted_values)
    plt.xlabel('Real Value')
    plt.ylabel('Predicted Value')
    plt.title(f'Real vs Predicted Values - {os.path.basename(csv_file)}')
    plt.grid()
    plt.legend()
    
    # Save the plot
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    plot_file = os.path.join(output_dir, f'{os.path.basename(csv_file).replace(".csv", ".png")}')
    plt.savefig(plot_file)
    plt.show()
```

```python
# Directory containing the CSV files
predictions_dir = 'data\predictions'
    
# Iterate over all CSV files in the directory
for csv_file in os.listdir(predictions_dir):
    if csv_file.endswith('.csv'):
        plot_predictions(os.path.join(predictions_dir, csv_file))
```

```python
def plot_predictions(csv_file, output_dir='ML_plots'):
    # Read the CSV file
    df = pd.read_csv(csv_file)

    # Extract the real and predicted values
    real_values = df['Real']
    predicted_values = df['Predicted']

    # Calculate the error
    error_values = real_values - predicted_values

    # Create a histogram for error values
    plt.figure(figsize=(10, 6))
    plt.hist(error_values, bins=30, range=(-100, 100))
    plt.xlabel('Error Value')
    plt.ylabel('Frequency')
    plt.title(f'Error Distribution - {os.path.basename(csv_file)}')
    plt.grid()
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    plot_file = os.path.join(output_dir, f'{os.path.basename(csv_file).replace(".csv", "_error_histogram.png")}')
    plt.savefig(plot_file)
    plt.show()

# Directory containing the CSV files
predictions_dir = 'data\\predictions'

# Iterate over all CSV files in the directory
for csv_file in os.listdir(predictions_dir):
    if csv_file.endswith('.csv'):
        plot_predictions(os.path.join(predictions_dir, csv_file))
```

```python
def plot_metrics(csv_file, output_dir='ML_plots'):
    # Read the CSV file
    df = pd.read_csv(csv_file)

    # Extract the MSE and MAE values
    mse_values = df['MSE']
    mae_values = df['MAE']
    model_names = df['Model']

    # Create a single plot with two y-axes
    fig, ax1 = plt.subplots(figsize=(10, 6))

    ax1.set_xlabel('Model')
    ax1.set_ylabel('MSE', color='tab:blue')
    mse_line, = ax1.plot(model_names, mse_values, marker='o', color='tab:blue', label='MSE')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    ax1.set_yscale('log')
    ax1.set_xticklabels(model_names, rotation=45, ha='right')

    ax2 = ax1.twinx()
    ax2.set_ylabel('MAE', color='tab:red')
    mae_line, = ax2.plot(model_names, mae_values, marker='x', color='tab:red', label='MAE')
    ax2.tick_params(axis='y', labelcolor='tab:red')
    ax2.set_yscale('log')

    fig.tight_layout()
    plt.title(f'MSE and MAE Values - {os.path.basename(csv_file)}')
    plt.grid()
    plt.legend(handles=[mse_line, mae_line])

    # Save the plot
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    plot_file = os.path.join(output_dir, f'{os.path.basename(csv_file).replace(".csv", "_mse_mae.png")}')
    plt.savefig(plot_file)
    plt.show()

# Plot the metrics for ranking.csv
plot_metrics('data/ranking.csv')
```
