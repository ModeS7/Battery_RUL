```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_predictions(csv_file, output_dir='ML_plots'):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Extract the real and predicted values
    real_values = df['Real']
    predicted_values = df['Predicted']
    
    # Create a plot
    plt.figure(figsize=(10, 6))
    plt.scatter(real_values, predicted_values)
    plt.xlabel('Real Value')
    plt.ylabel('Predicted Value')
    plt.title(f'Real vs Predicted Values - {os.path.basename(csv_file)}')
    plt.grid()
    plt.legend()
    
    # Save the plot
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    plot_file = os.path.join(output_dir, f'{os.path.basename(csv_file).replace(".csv", ".png")}')
    plt.savefig(plot_file)
    plt.show()
```

```python
# Directory containing the CSV files
predictions_dir = 'data\predictions'
    
# Iterate over all CSV files in the directory
for csv_file in os.listdir(predictions_dir):
    if csv_file.endswith('.csv'):
        plot_predictions(os.path.join(predictions_dir, csv_file))
```
